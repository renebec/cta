```python
import pandas as pd
import plotly.express as px
from IPython.display import display, HTML
```

```python


# 1. Create a sample DataFrame
data = {
    'NAME': ['Pedro', 'María', 'Bob', 'Eve', 'Tom', 'Sara'],
    'AGE': [29, 18, 22, 29, 35, 27],
    'CITY': ['Oaxaca', 'Monterrey', 'Chicago', 'Miami', 'Dallas', 'Austin']
}
df = pd.DataFrame(data)

# 2. Extract the first row from the DataFrame
first_row = df.iloc[0]

# 3. Read the existing HTML template from 'index.html'
with open('index.html', 'r') as file:
    html_content = file.read()

# 4. Replace placeholders with values from the first row
html_content = html_content.replace("{{name}}", first_row['NAME'])
html_content = html_content.replace("{{age}}", str(first_row['AGE']))
html_content = html_content.replace("{{city}}", first_row['CITY'])

# 5. Save the interactive histogram using Plotly
fig = px.histogram(df, x="AGE", nbins=5, title="Age Distribution")
fig.write_html("histogram.html")  # Save the Plotly figure as an HTML file

# 6. Convert the DataFrame to an HTML table (with Jupyter-like style using Bootstrap)
df_html = df.to_html(classes='table table-bordered table-hover', index=False)

# 7. Embed the interactive table into the HTML content
html_content = html_content.replace("{{df_table}}", df_html)

# 8. Embed the histogram inside the <iframe> tag
html_content = html_content.replace("<iframe src=\"histogram.html\" width=\"600\" height=\"400\"></iframe>", 
                                    f'<iframe src="histogram.html" width="600" height="400" style="border:none;"></iframe>')

# 9. Save the final updated HTML file
with open('index_final.html', 'w') as file:
    file.write(html_content)

# 10. Display the final HTML content in Jupyter
display(HTML(html_content))

print("HTML file has been updated and saved as 'index_final.html'.")
```
